```python
#Fetch the data
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Start the web driver and open the web page
driver = webdriver.Firefox()
driver.get('https://qrco.de/beV4dT')  # Replace with the URL of the web page you want to scrape

# Find and retrieve all tables
tables = driver.find_elements(By.TAG_NAME, 'table')

table_data = {}  # Dictionary to store the table data

for table in tables:
    # Get the ID attribute of the table, or use a default name if it doesn't have an ID
    table_id = table.get_attribute('id')
    if not table_id:
        table_id = f"table_{len(table_data) + 1}"

    # Parse the HTML table with BeautifulSoup
    table_html = table.get_attribute('outerHTML')
    df = pd.read_html(table_html)[0]  # Assuming there's only one table per page, change the index if needed

    # Store the DataFrame in the dictionary with the table's ID as the key
    table_data[table_id] = df

# Clean up and close the web driver
driver.quit()

# Now you can work with the dictionary where keys are table IDs and values are DataFrames
for table_id, df in table_data.items():
    print(f"Table ID: {table_id}")
    print(df)
```

```python
pd.DataFrame([table_data]).to_json('recensement.json')
```

```python
#Fetch the graphical representations 
from selenium import webdriver
import time

# URL of the web page with animated SVG images
url = "https://qrco.de/beV4dT"  # Replace with the URL of the web page

# Initialize a headless Chrome browser
driver = webdriver.Firefox()
# Navigate to the web page
driver.get(url)

# Wait for the page to load (adjust the wait time as needed)
time.sleep(30)

# Find and retrieve all SVG elements with animations
svg_elements = driver.find_elements(By.TAG_NAME, 'svg')

# Download and save each animated SVG with its animation
for i, svg_element in enumerate(svg_elements):
    # Generate a unique filename for each SVG
    filename = f"animated_svg_{i}.svg"

    # Get the inner HTML of the SVG element
    svg_inner_html = svg_element.get_attribute("innerHTML")

    # Save the SVG to a file
    with open("graphics/"+filename, "w", encoding="utf-8") as svg_file:
        svg_file.write(f'<svg xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink">{svg_inner_html}</svg>')

    print(f"Saved {filename}")

# Close the browser
driver.quit()
```
